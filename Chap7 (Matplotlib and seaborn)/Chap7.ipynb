* Matplotlib: Là thư viện cơ bản nhất để vẽ đồ thị trong Python. Nó cung cấp một loạt các hàm để tạo ra các loại biểu đồ khác nhau, từ đơn giản đến phức tạp.
* Seaborn: Được xây dựng trên nền tảng Matplotlib, Seaborn cung cấp một giao diện cấp cao hơn, giúp chúng ta tạo ra những biểu đồ đẹp mắt và mang tính thống kê cao một cách dễ dàng hơn.  

* Line	plot()
* Scatter	scatter()
* Bar	bar(), barh()
* Histogram	hist()
* Pie Chart	pie()
* 3D Plot	plot_surface(), scatter3D()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
fig, axs = plt.subplots(3, 2, figsize=(10, 6))
axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].bar(range(5), [3, 7, 1, 8, 5])
axs[1, 1].hist(np.random.randn(1000), bins=30)
axs[2, 0].pie([1, 2, 3, 4], labels=['A', 'B', 'C', 'D'])
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(0, 2 * np.pi, 50)
r = np.linspace(0, 5, 50)

T, R = np.meshgrid(theta, r) # t ma tran goc , r ma tran ban kinh tuong ung voi t
X = R * np.cos(T)
Y = R * np.sin(T)
Z = R 

ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_title("Hình nón 3D")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('sales_data.csv')

# Vẽ biểu đồ Sales theo Date
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Sales'], color='blue')
plt.title('Sales by Date')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

**Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dữ liệu ví dụ
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Sales': [100, 200, 150, 300],
    'Revenue': [400, 600, 500, 700],
    'Profit': [50, 150, 100, 200]
}

df = pd.DataFrame(data)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
sns.barplot(x='Category', y='Sales', data=df, palette='Blues', ax=axs[0, 0])
axs[0, 0].set_title('Sales by Category')
axs[0, 0].set_xlabel('Category')
axs[0, 0].set_ylabel('Sales')

sns.scatterplot(x='Sales', y='Revenue', data=df, color='green', ax=axs[0, 1])
axs[0, 1].set_title('Sales vs Revenue')
axs[0, 1].set_xlabel('Sales')
axs[0, 1].set_ylabel('Revenue')

sns.boxplot(x='Category', y='Profit', data=df, palette='coolwarm', ax=axs[1, 0])
axs[1, 0].set_title('Profit by Category')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Profit')

sns.histplot(df['Sales'], kde=True, color='purple', ax=axs[1, 1])
axs[1, 1].set_title('Distribution of Sales')
axs[1, 1].set_xlabel('Sales')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```
