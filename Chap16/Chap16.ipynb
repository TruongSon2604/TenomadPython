**Dự đoán với RMSE và MSE trong Machine Learning (ML)**
* Trong học máy (ML), đặc biệt là trong các bài toán hồi quy (regression), việc đánh giá độ chính xác của mô hình là rất quan trọng. Một trong những chỉ số phổ biến được sử dụng để đo lường độ chính xác của mô hình hồi quy là RMSE (Root Mean Squared Error) và MSE (Mean Squared Error).
1. MSE (Mean Squared Error)
* MSE là chỉ số đo lường độ sai lệch bình quân giữa giá trị dự đoán và giá trị thực tế. Nó tính toán bình phương của sự khác biệt giữa giá trị thực tế và giá trị dự đoán của mô hình, sau đó lấy trung bình của các bình phương này.
* 
* yi : là giá trị thực tế
* yi^: là giá trị dự đoán
* n : là sô lượng mẫu trong bộ dữ liệu
* Ý nghĩa:
* MSE càng nhỏ, mô hình càng chính xác.

**2. RMSE (Root Mean Squared Error)**
RMSE là căn bậc hai của MSE, giúp đưa chỉ số này về cùng đơn vị với dữ liệu gốc, giúp dễ dàng giải thích hơn. RMSE đo lường mức độ sai lệch trung bình giữa các giá trị dự đoán và giá trị thực tế. Nó cho biết mô hình có sai lệch bao nhiêu so với dữ liệu thực tế.
* 
* Ý nghĩa:
* RMSE có cùng đơn vị với giá trị dự đoán và giá trị thực tế, giúp dễ dàng hiểu và giải thích.
* Giống như MSE, RMSE càng nhỏ thì mô hình càng chính xác.
* Ưu điểm: RMSE dễ hiểu hơn vì nó có cùng đơn vị với dữ liệu gốc. RMSE cũng nhạy cảm với các giá trị ngoại lai (outliers).

các hàm accuracy_score, precision_score, recall_score, và f1_score chủ yếu được sử dụng cho các bài toán phân loại (classification), không phải cho các bài toán hồi quy (regression).

Accuracy, Precision, Recall, F1-Score đều được thiết kế để đánh giá mô hình phân loại, đặc biệt là trong các bài toán phân loại nhị phân hoặc đa lớp. Chúng đều yêu cầu các giá trị dự đoán (y_pred) và giá trị thực tế (y_test) phải là các nhãn phân loại (ví dụ: 0 hoặc 1 cho phân loại nhị phân).

Cách thức hoạt động trong bài toán phân loại:
Accuracy: Tính tỷ lệ dự đoán đúng so với tổng số mẫu.

Precision: Tỷ lệ giữa số dự đoán đúng là positive trên tổng số dự đoán là positive.

Recall: Tỷ lệ giữa số dự đoán đúng là positive trên tổng số thực tế là positive.

F1-Score: Trung bình hài hòa giữa Precision và Recall.

Tuy nhiên, nếu bạn đang làm việc với một bài toán hồi quy, các metric này không còn phù hợp, vì hồi quy liên quan đến việc dự đoán các giá trị liên tục, không phải các nhãn phân loại.
